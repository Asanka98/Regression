```python
def cost_function(X, Y, w, b):

    cost = 0

    m = X.shape[0]

    for i in range(m):

            y_wb = X[i] * w + b

            cost_i = ( y_wb - Y[i] ) ** 2
            
            cost += cost_i

    return cost / ( 2 * m )
```

```python
print("overall cost : " + str(cost_function(X_train, Y_train, 2.5, 21.8)))
```
